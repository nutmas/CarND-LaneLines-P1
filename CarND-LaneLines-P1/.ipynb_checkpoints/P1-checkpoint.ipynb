## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math
    
def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    
    
    # convert from RGB as read in with imread()
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 

def other_masking(image):
    
    # make HLS & HSV image
    HSL_image = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
    HSV_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    
    #HLS_image = hls_image(image)
    #set colour thresholds for detection
    sensitivity = 30
    #white mask thresholds
    white_lower_threshold = np.uint([0,0,255-sensitivity])
    white_upper_threshold = np.uint([255,sensitivity,255])
    #yellow mask thresholds
    yellow_lower_threshold = np.uint([20,120,80])
    yellow_upper_threshold = np.uint([45,200,255])
    # create masks for each colour
    # white mask on HSV image
    white_mask = cv2.inRange(HSV_image,white_lower_threshold,white_upper_threshold)
    # white mask on HSL image
    yellow_mask = cv2.inRange(HSL_image,yellow_lower_threshold,yellow_upper_threshold)
    
    # combine to make one mask
    final_mask = cv2.bitwise_or(white_mask, yellow_mask)
    # return mask of original image
    return cv2.bitwise_and(image, image, mask=final_mask)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)
    
    
def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with - black image
    mask = np.zeros_like(img) 
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    #plt.imshow(masked_image)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=3):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    
    # using image co-ordinate system x: left to right, y: top to bottom
    # low x, high y: high x, low y = left lane, negative slope
    # low x, low y: high x, high y = right lane, positive slope
    
    # lists to hold left and right lines
    left_line = []
    right_line = []
    left_lengths = []
    right_lengths = []
    
    # seperate lines into left or right groupings
    for line in lines:
        for x1,y1,x2,y2 in line:
            # calculate slope and intercept
            m = (y2-y1)/(x2-x1)
            # line equation y = m(slope)x + b(y intercept)
            b = y1 - m * x1
            #print("m: {}, b: {}".format(m,b))
            line_length = np.sqrt((y2-y1)**2+(x2-x1)**2)
            # seperate lines into left or right
            
            #right line
            if (y2-y1)/(x2-x1) > 0: #(m > 0.2 or m < 0.8): # and line_length > 10:
                right_line.append((m,b))
                right_lengths.append(line_length)
                #print ("right: {},{},{},{}".format(x1,y1,x2,y2))
            # left line
            elif (y2-y1)/(x2-x1) < 0: #(m > -0.8 or m < -0.2): # and line_length > 10:
                # limit variance of slope
                left_line.append((m,b))
                left_lengths.append(line_length)
                #print ("left: {},{},{},{}".format(x1,y1,x2,y2))
            else:
                # vertical line so ignore it
                continue

    
    
    left_lane  = np.dot(left_lengths,  left_line) /np.sum(left_lengths)  if len(left_lengths) >0 else None
    right_lane  = np.dot(right_lengths,  right_line) /np.sum(right_lengths)  if len(right_lengths) >0 else None
    
    #print ("left lane: ", left_lane)
    #print ("right lane: ", right_lane)
    
    
    # calculate mean of lists  
    left_mean = [sum(y)/len(y) for y in zip(*left_line)]
    right_mean = [sum(y)/len(y) for y in zip(*right_line)]
    #height of image to get max y value
    imshape = img.shape 
    y1 = imshape[0]
    y2 = 335 # fixed horizon point matching parameter defined in road lane mask
    
    # line equation x = (y - b)/m
    left_x1 = int((y2 - left_mean[1])/left_mean[0])
    left_y1 = int(y2)
    left_x2 = int((y1 - left_mean[1])/left_mean[0])
    left_y2 = int(y1)
    right_x1 = int((y1 - right_mean[1])/right_mean[0])
    right_y1 = int(y1)
    right_x2 = int((y2 - right_mean[1])/right_mean[0])
    right_y2 = int(y2)
    
    #print("left line: {}; {}; {}; {}".format(left_x1,left_y1,left_x2,left_y2))
    #print("right line: {}; {}; {}; {}".format(right_x1,right_y1,right_x2,right_y2))
    
    # generate annotated lines for left and right lines
    cv2.line(img, (left_x1, left_y1), (left_x2, left_y2), color, 15)
    cv2.line(img, (right_x1, right_y1), (right_x2, right_y2), color, 15)
    

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )

# helper function to show images
def show_image(sub_image, image_title, output_image):
    plt.subplot(*sub_image)
    plt.title(image_title)
    # if image is 3 channels
    if len(output_image.shape) == 3:
        # show colour image
        plt.imshow(output_image)
    else:
        #image is greyscale
        plt.imshow(output_image, cmap='gray')
````

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

````python
#Pipeline that will draw lane lines on test_images
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
import os
from datetime import datetime


# specify location of image files
images_dir = "test_images/"
# set path to save test images to
path = "test_images_output/"

# loop through images in a directory
for file in os.listdir(images_dir):
    # if image is found, run line detect
    if file.endswith(".jpg"):

        # read in image from directory
        img = mpimg.imread(os.path.join(images_dir,file))

        # convert image to grey scale
        gray = grayscale(img)
        
        # apply guassian to blur grayscale image
        blur_gray = gaussian_blur(gray, 5)
        #blur_colour = gaussian_blur(gray, 3)
        
        # apply canny transform and set parameters - omitted blur; apply canny direct to masked coloured lines
        edges = canny(blur_gray, 50, 150)

        # create vertices for lane area
        imshape = img.shape
        vertices = np.array([[(30, imshape[0]), (450, 325), (530, 325), (imshape[1] - 20, imshape[0])]], dtype=np.int32)
        
        # create mask to focus area of line detection
        masked_edges = region_of_interest(edges, vertices)
        
        # apply hough on masked image; set rho, theta, threshold, min line length, and max gap between connections
        line_image = hough_lines(masked_edges, 3, np.pi / 180, 10, 40, 20)

        # Draw the lines on the edge image
        lines_edges = weighted_img(line_image, img, =0.8, =1., =0.)

        # set the output focus
        result = lines_edges
        
        plt.figure(figsize=[8, 6])
        plt.imshow(result)
        plt.title(file)
        
        plt.figure(figsize=[16, 9])
        for i, image_name in enumerate(['img', 'gray', 'blur_gray', 'edges', 'masked_edges', 'line_image']):
            show_image((3, 2, i+1), image_name, eval(image_name))

        # save each to the test_images_output directory; correct the colours back to BGR for file write
        cv2.imwrite(os.path.join(path ,"FinalOutput_"+file),cv2.cvtColor(result, cv2.COLOR_RGB2BGR))
````

````python
def process_image(image):
    
    # convert image to grey scale
    gray = grayscale(image)

    # apply guassian to blur grayscal image
    blur_gray = gaussian_blur(gray, 5)
        
    # apply canny transform and set parameters - omitted blur; apply canny direct to masked coloured lines
    edges = canny(blur_gray, 50, 150)

    # create vertices for lane area
    imshape = image.shape
    vertices = np.array([[(0, imshape[0]), (450, 325), (530, 325), (imshape[1], imshape[0])]], dtype=np.int32)

    # create mask to focus area of line detection
    masked_edges = region_of_interest(edges, vertices)
    
    # apply hough on masked image; set rho, theta, threshold, min line length, and max gap between connections
    line_image = hough_lines(masked_edges, 3, np.pi / 180, 10, 40, 20)

    # Draw the lines on the edge image
    lines_edges = weighted_img(line_image, image, =0.8, =1., =0.)

    # set the output focus
    result = lines_edges

    return result
````

````python
def colour_masking(image):
    
    # make a copy to avoid altering master
    local_image = np.copy(image)
    # make HLS image
    hsv_image = cv2.cvtColor(local_image, cv2.COLOR_RGB2HSV)
    # create masks for each colour
    #colour mask with thresholds matched for yellow
    yellow_mask = cv2.inRange(hsv_image, (20, 80, 80), (25, 255, 255))
    #colour mask with thresholds matched for white
    sensitivity = 15
    white_mask = cv2.inRange(hsv_image, (0, 0, 180), (255, 25, 255))
    # combine to make one mask
    return cv2.bitwise_or(yellow_mask, white_mask)  
````
